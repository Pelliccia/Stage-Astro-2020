```python
import numpy as np
import copy
```

```python
def PhasesGenerator(Datemid,Endmid):
    Date=Datemid
    x=[0., 0.0625, 0.1875, 0.25, 0.3125, 0.375, 0.4375, 0.5, 0.5625, 0.625, 0.6875, 0.75, 0.8125, 0.875, 0.9375]
    phases=[]
    phase=[]
    for i in range(0,15):
        while Date<=Endmid:
            phase.append(Date+x[i]*0.407521)
            Date=Date+0.407521
        Date=Datemid
        phases.append(phase)
        phase=[]
    return(phases)
```

```python
PG=PhasesGenerator(2458933.278,2458975.322) #Va du mid du 25/03 au 06/05
```

```python
def PhasesFiltred(PG):
    PGc=copy.deepcopy(PG)
    i, j = 0, 0                    #Array ou liste, cette partie-là est identique jusqu'au "while tampon!=[]"
    L, l = len(PG), len(PG[i])
    tampon,lt = [], []
    for i in range(0,L):
        for j in range(0,l): 
            if PG[i][j] >= 2458945. and PG[i][j]<= 2458948.:
                tampon.append(PG[i][j])
            elif ((PG[i][j])%1) <= 0.6455 or ((PG[i][j])%1) >= 0.9375:
                tampon.append(PG[i][j])
        lt.append(len(tampon))
        for k in range(0,l):
            if lt[i]==0:
                    break
            if tampon[0]==PGc[i][k]: #print(PGc[i][j],tampon[0]) #détecte bien la correspondance mais remove ne se fait pas...
                PGc[i][k]=0 
                tampon.remove(tampon[0])
                
        while 0 in PGc[i]:
            PGc[i].remove(0)
    return(PGc)
```

```python
def ConverterJD(jul): #Convertisseur pour une date précise
    Z=int(jul+0.5)
    F=(jul+0.5)%1
    a=int((Z-1867216.25)/36524.25)
    A=Z+1+a-int(a/4)
    B=A+1524
    C=int((B-122.1)/365.25)
    D=int(365.25*C)
    E=int((B-D)/30.6001)
    d=int(B-D-int(30.6001*E)+F)
    if E<14:
        m=E-1
    else:
        m=E-13
    if m>2:
        y=C-4716
    else:
        y=C-4715
    hour=abs(24*F)
    mini=(hour-int(hour))*60
    sec=(mini-int(mini))*60
    return(m,d,y,int(hour),int(mini),sec)

def TimeConverter(PF): #Convertisseur pour une liste entrante ou une array entrante
    PFC=copy.deepcopy(PF)
    x=[0., 0.0625, 0.1875, 0.25, 0.3125, 0.375, 0.4375, 0.5, 0.5625, 0.625, 0.6875, 0.75, 0.8125, 0.875, 0.9375]
    for i in range(0,len(PFC)):
        for j in range(0,len(PFC[i])):
            PFC[i][j]=ConverterJD(PF[i][j])
        print("\n Phase:",x[i],"\n",PFC[i])
        j=0
```
